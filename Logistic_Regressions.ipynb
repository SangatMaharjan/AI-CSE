```python
# Loading the Dataset
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
X=data.data
Y=data.target
```

```python
# Exploring the Dataset

# Display the shape of X and Y

import numpy as np

print(f"Shape of X : {X.shape}")
print(f"Shape of Y: {Y.shape}")

# Print the first five rows
print(f"First Five Rows : {X[:5]}")

# Identify the number of samples , features and classes

print(f"Number of samples : {X.shape[0]}")
print(f"Number of features : {X.shape[1]}")
print(f"Numer of classes : {len(np.unique(Y))}")
```

```python
# Splitting the Dataset into 80% Training and 20% Testing

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.20,      # 0.20 means 20% goes to the testing set
    random_state=42      # Ensures the split is identical every time you run it
)

#Verify the split sizes
print(f"Training features shape (X_train): {X_train.shape}")
print(f"Testing features shape (X_test):  {X_test.shape}")

```

```python
# Standardize Features

from sklearn.preprocessing import StandardScaler
#Initialize the StandardScaler
scaler = StandardScaler()
#Fit on training data and transform it

X_train_scaled = scaler.fit_transform(X_train)
# Transform the testing data (using the training data's scale)
X_test_scaled = scaler.transform(X_test)

```

```python
# Train Logistic Regression Model

from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model with default parameters
model = LogisticRegression()

# Train (fit) the model using the scaled training data
model.fit(X_train_scaled, Y_train)

print("Logistic Regression model trained successfully!")
```

```python
# Make Predictions

# Predicting class labels for test set

# Use the trained model to predict the class labels for the test set
Y_pred = model.predict(X_test_scaled)

# Printing out the first 10 predictions
print("First 10 predicted labels: ", Y_pred[:10])
print("First 10 actual labels:    ", Y_test[:10])
```

```python
# Evaluate the Model

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Computing the evaluation metrics
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
cm = confusion_matrix(Y_test, Y_pred)

print(f"Accuracy Score : {accuracy} ")
print(f"Precision Score : {precision}")
print(f"Recall Score : {recall}")
print(f"F1 Score : {f1}")
print(f"Confusion Matrix :\n{cm}")

```
