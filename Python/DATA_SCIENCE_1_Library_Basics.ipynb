```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Define our dataset
dataset = [10,11,12,13,14,15,16,17,18,20,22,24,25,26,30,34,32,33,40,45,50,55,60,90,120,180]
```

```python
sns.histplot(dataset,color = 'r', kde = True) #kde= kernal density function
```

```python
sns.histplot(dataset)
```

```python
## Z score
outliers =[]
#function to take data as argument and return possible outliers
def detect_outliers(data):
    threshold =3 # here 3 is standard deviation
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```

```python
# IQR ( InterQuartile Range)
# 1. Sort the data
# 2. Calculate Q1 and Q3
# 3. IQR = (Q3 - Q1 )
# 4. Find the Lower fence(q1 - 1.5(iqr))
# 5. Find upper fence(q3 + 1.5(iqr))
```

 IQR: the interquartile range tells you the spread of the middle half of your distribution. Quartiles segment any distribution that's ordered from low to high into four equal parts. The interquartile range (IQR) contains the second and third quartiles, or the middle half of your data set.

```python
dataset = sorted(dataset)
#data set
```

```python
q1,q2,q3 = np.percentile(dataset,[25,50,75])
#(q1=median of upper half)
#(q3=median of lower half)
#(q2=median of whole table)
```

```python
print(q1,q2,q3)
```

```python
iqr = q3-q1
print(iqr)
```

upper and lower fences represent the cut-off values for upper and lower outliers in a dataset

```python
#Find the Lower fence and higher fence
lower_fence=q1-(1.5*iqr)
higher_fence=q3+(1.5*iqr)
```

```python
print(lower_fence,higher_fence)
```

```python
#Remove the outliers which doesnot belongs to lower -->higher fence range
outliers =[]
for item in dataset:
    if item<lower_fence or item>higher_fence:
        outliers.append(item)
```

```python
print(outliers)
```
