```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50, 51, 52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weakly Weather')
plt.plot(days, max_t, label = "max")
plt.plot(days, min_t, label = "max")
plt.plot(days, avg_t, label = "average")
plt.legend(loc='best') #to display the labels
#plt.savefig("Sample_Weeklyweather.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#_____________Simple Bar chart showing revenues of majrt US tech companies_____________
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90, 136, 89, 27]
```

```python
xpos =  np.arange(len(company))
print(xpos)
```

```python
plt.bar(xpos,revenue,label='Revenue',color='green')


plt.xticks(xpos,company)
plt.xlabel('Tech companies')
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
profit = [40,2,34,12]
```

```python
plt.figure(figsize=(30,10))
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.xlabel("Tech Companies")
plt.title('US Technology Stocks')
plt.legend()
```

```python
#_________________________Horizontal bar chart___________________________________________
```

```python
plt.barh(xpos,revenue,label='Revenue')

plt.yticks(xpos,company)
plt.xlabel('Revenue(Bln)')
plt.ylabel("Tech Companies")
plt.title('US Technology Stocks')
plt.legend()
```

```python
#________________________PIE GRAPH_____________________________________________________
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels);
```

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels,shadow=True, autopct='%1.2f%%', radius=2)
plt.show()
```

```python
#_____________EXPLODE___________________
```

```python
plt.pie(exp_vals, labels=exp_labels,shadow=True, autopct='%1.2f%%', radius=2, explode=[0,0,0,0.5,0.7])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,shadow=True, autopct='%1.2f%%', radius=2, explode=[0,0,0,0.5,0.7],counterclock=True,startangle=45)
plt.show()
```
