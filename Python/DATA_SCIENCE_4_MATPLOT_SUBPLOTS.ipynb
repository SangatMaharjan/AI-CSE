```python
import matplotlib.pyplot as plt
```

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expense=[2800,3000,2800,3400]

plt.subplot(2,1,1)# (row, col, position)
plt.title('Income and Expense')
plt.xlabel("Year")
plt.ylabel("Income")
plt.bar(year,income,color='red',label='income')
plt.legend()

plt.subplot(2,1,2)
plt.bar(year,expense,color='green',label='expense')
plt.xlabel("Year")
plt.ylabel('Expense')
plt.legend()

```
